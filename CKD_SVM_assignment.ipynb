```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=.30,random_state=0)
```

```python
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV
param_grid = {'kernel': ['linear','rbf','poly','sigmoid'], 
              'gamma': ['auto','scale'], 
              'C':[10,100,500,1000,2000,5000]}
grid=GridSearchCV(SVC(probability =True), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,grid_predictions)
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average="weighted")
print(f1_macro)
print("The f1_score value for best parameter{}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test) [:,1])
```
